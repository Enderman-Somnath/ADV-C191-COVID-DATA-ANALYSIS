```python
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()
df
```

```python
#Task 1 
#Sort the data as per total number of cases
sorted_df = df.sort_values(by=['total_cases'])
sorted_df
```

```python
#Task 2
#Get top 5 countries who has the least number of cases and plot a bar graph
least_cases_ = sorted_df['total_cases'].head(5)
print(least_cases_)

least_cases_country = sorted_df['country'].head(5)
print(least_cases_country)

plt.xlabel("country")
plt.xticks(rotation='vertical')
plt.ylabel("covid_cases")

label = least_cases_country
value = least_cases_
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue'))
```

```python
#Task 3
#Sort the data as per total number of deaths

sorted_df = df.sort_values(by=['total_deaths'])
sorted_df
```

```python
#Task 4
#Get top 5 countries who has the maximum number of deaths and plot a bar graph
highest_death_ = sorted_df['total_deaths'].tail(5)
print(highest_death_)

highest_death_country = sorted_df['country'].tail(5)
print(highest_death_country)

plt.xlabel("country")
plt.xticks(rotation='vertical')
plt.ylabel("covid_cases")

label = highest_death_country
value = highest_death_
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue'))

```

```python
#Task 5
#Sort the data as per active cases
sorted_df = df.sort_values(by=['active_cases'])
sorted_df
```

```python
#Task 6
#Get top 5 countries who has the maximum number of active cases and plot a bar graph
highest_death_ = sorted_df['active_cases'].tail(5)
print(highest_death_)

highest_death_country = sorted_df['country'].tail(5)
print(highest_death_country)

plt.xlabel("country")
plt.xticks(rotation='vertical')
plt.ylabel("covid_cases")

label = highest_death_country
value = highest_death_
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue'))
```
